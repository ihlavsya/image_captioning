```python
%matplotlib inline
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
import numpy as np
import skimage.io as io
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
prefix = 'captions'
```

```python
dataDir='dataset'
dataType='val2017'
annFile = '%s/annotations/%s_%s.json'%(dataDir,prefix,dataType)
cocoGt=COCO(annFile)
```

```python
imgIds = cocoGt.getImgIds()
imgIds=imgIds[0:3]
print(imgIds)
```

```python
import os
```

```python
dir_path = 'dataset/val2017'
extension = '.jpg'
base_number = 1000000000000
images_paths = []
for img_id in imgIds:
    number = base_number + img_id
    filename = str(number)[1:] + extension
    image_path = os.path.join(dir_path, filename)
    images_paths.append(image_path)
```

```python
print(images_paths)
```

```python
for i in range(3):
    annId = cocoGt.getAnnIds(imgIds=[imgIds[i]])
    ann = cocoGt.loadAnns(annId)
    cocoGt.showAnns(ann)
    # img = cocoGt.loadImgs(imgIds[i])
    I = io.imread(images_paths[i])
    plt.axis('off')
    plt.imshow(I)
    plt.show()
    print('--------------------')
```
